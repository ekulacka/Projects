```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
 
data.head()
```

```python
print(data.groupby('class').sum())
```

```python
y = data["class"]
x = data.drop(["class"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
y_predict = lr_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, roc_curve, precision_score, roc_auc_score, f1_score

print("Accuracy is: "+ str(round(lr_model.score(x_test,y_test),3)))
print("Recall score is: " + str(round(recall_score(y_test, y_predict),3)))
print("Precision score is: " + str(round(precision_score(y_test, y_predict),3)))
print("F1 score is: " + str(round(f1_score(y_test, y_predict),3)))
print("\nConfusion matrix:")
print(confusion_matrix(y_test, y_predict))
```

```python
from sklearn.preprocessing import LabelEncoder
X = data.drop(["class"], axis=1)
y = data["class"]

X = pd.get_dummies(X)

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
y_predict = lr_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, roc_curve, precision_score, roc_auc_score, f1_score

print("Accuracy is: "+ str(round(lr_model.score(X_test,y_test),3)))
print("Recall score is: " + str(round(recall_score(y_test, y_predict),3)))
print("Precision score is: " + str(round(precision_score(y_test, y_predict),3)))
print("F1 score is: " + str(round(f1_score(y_test, y_predict),3)))
print("\nConfusion matrix:")
print(confusion_matrix(y_test, y_predict))
```
